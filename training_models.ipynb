```python
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
column_names = np.array(["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation","relationship", "race", "sex", "capital-gain", "capital-loss", "hours-pr-week", "native-country", "income"])
catagorical = np.array(["workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation","relationship", "race", "sex", "native-country"])
data = pd.read_csv('data/adult.data', nrows=1000, header=None, names=column_names)
data = pd.get_dummies(data, columns=catagorical)

X = data.drop(columns=["income"]).astype("float32").to_numpy()
y = data["income"].to_numpy()
```

```python
col_names = data.columns
```

```python
#decition tree
criterion = "gini" # kunne også være "entropy"
dtc = tree.DecisionTreeClassifier(criterion=criterion, min_samples_split=2, max_depth = 5)
dtc = dtc.fit(X, y)
```

```python
fname = "tree" + criterion + ".png"

fig = plt.figure(figsize=(4, 4), dpi=200)
_ = tree.plot_tree(dtc, filled=False, feature_names=col_names)
plt.savefig(fname)
plt.show()
```
